### Import necessary liabraries

```python
import pandas as pd
import numpy as np
```

### Load the file created in part1 and create a second dataframe using the file for latitudes and longitutudes

```python
df_toronto = pd.read_csv('toronto_part1.csv', index_col=[0])
df_toronto.head()
```

```python
df_ll = pd.read_csv('Geospatial_Coordinates.csv')
df_ll.rename(index=str, columns={"Postal Code": "Postcode"}, inplace=True)
df_ll.head()
```

### Join the two dataframes based on postcode

```python
df_toronto2 = pd.merge(df_toronto, df_ll, on='Postcode')
df_toronto2.to_csv('toronto_part2.csv')
```
